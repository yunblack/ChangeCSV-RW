```python
import csv
```

```python
AREA_CODE=[]
AREA_NAME=[]
TEMPERATURE=[]
HUMIDITY=[]
DATE=[]
RAIN=[]
SNOW=[]
CLOUD=[]
PRESSURE=[]
WIND=[]
PAIN=[]

MONTH=[]
DAY=[]
HOUR=[]

NEWDATE=[]
NEWNEWDATE=[]

# 지역 코드 (AREA_CODE)
# 지역 이름 (AREA_NAME)
# 온도 (TEMPERATURE)
# 습도 (HUMIDITY)
# 날짜 (DATE)
# 강수량 (RAIN)
# 적설량 (SNOW)
# 전운량 (CLOUD)
# 기압 (PRESSURE)
# 풍속 (WIND)
# 불쾌지수 (PAIN)

def readCsv():
    with open('DATASETS.csv', mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                line_count += 1
                  
            if row["DATE"] != None:        
                DATE.append(row["DATE"])
           
            if row["LOCATION"] != None:
                AREA_NAME.append(row["LOCATION"])
                  
            if row["TEMP"] != None:
                TEMPERATURE.append(convertToFloat(row["TEMP"]))
                  
            if row["HUMID"] != None:
                HUMIDITY.append(convertToFloat(row["HUMID"])) 
                  
            if row["RAIN"] != None:
                RAIN.append(convertToFloat(row["RAIN"]))
            else:
                RAIN.append('0')
                  
            if row["SNOW"] != None:
                SNOW.append(convertToFloat(row["SNOW"]))
            else:
                SNOW.append('0')
                  
            if row["WIND"] != None:
                WIND.append(convertToFloat(row["WIND"]))
            else:
                WIND.append('0')
                  
            if row["PRES"] != None:
                PRESSURE.append(convertToFloat(row["PRES"]))
                  
            if row["CLOUD"] != None:
                CLOUD.append(convertToFloat(row["CLOUD"]))
            else:
                CLOUD.append("0")
                  
            if row["AREA_CODE"] != None:
                AREA_CODE.append(convertToFloat(row["AREA_CODE"]))
                  
            line_count += 1
    
    checkblank()
    calHumidity()
    calWindChill()
    splitDate()
    
    makeNewDateFormat()
    
    print('Reading CSV -- Completed')
    
def splitDate():
    temp=''
    for i in range (0, DATE.__len__()):
        
        if str(((DATE[i].split('-')[2]).split()[1]).split(':')[0]).__len__()== 1:
            temp='' + str(DATE[i].split('-')[1]) + str((DATE[i].split('-')[2]).split()[0]) + str('0') + str(((DATE[i].split('-')[2]).split()[1]).split(':')[0]) + ''
        else:
            temp='' + str(DATE[i].split('-')[1]) + str((DATE[i].split('-')[2]).split()[0]) + str(((DATE[i].split('-')[2]).split()[1]).split(':')[0]) + ''        
        
        NEWNEWDATE.append(temp)
        # print(temp)
        
        MONTH.append(float(DATE[i].split('-')[1]))
        DAY.append(float((DATE[i].split('-')[2]).split()[0]))
        HOUR.append(float(((DATE[i].split('-')[2]).split()[1]).split(':')[0]))
        
        #print(i,MONTH[i],DAY[i],HOUR[i])

def makeNewDateFormat():
    temp=''
    for i in range (0, DATE.__len__()):
        temp= str(int(MONTH[i]))+'-'+str(int(DAY[i]))+' '+str(int(HOUR[i]))+':00'
        # print(temp)
        NEWDATE.append(temp)
        
        
def checkblank():
    for i in range (0, DATE.__len__()):
        if CLOUD[i]=='':
            CLOUD[i]='0'
        if DATE[i]=='':
            DATE[i]='0'
        if RAIN[i]=='':
            RAIN[i]='0'
        if SNOW[i]=='':
            SNOW[i]='0'
        if AREA_NAME[i]=='서울':
            AREA_NAME[i]='Seoul'
          

def calHumidity():
    
    for i in range (0,DATE.__len__()):
        try:  
            HUMIDITY[i]=float(HUMIDITY[i])/100
        except ValueError:
            HUMIDITY[i]=0
            print("Error", DATE[i])
    
def convertToFloat(data):
    
    newdata=0    
    try:  
        newdata=float(data)
    except ValueError:
        newdata=0    
    return newdata

def calWindChill():
    
    temp = 0
    
    for i in range (0,DATE.__len__()):
        try:
            temp = (float(TEMPERATURE[i]) * 1.8) - 0.55 * (1-float(HUMIDITY[i])) * (1.8*float(TEMPERATURE[i]) - 26) + 32  
            PAIN.append(float(temp))
        
        except ValueError:
            PAIN.append(0)
            print("Error", DATE[i])
```

```python
def writeCsv():
    with open('FIRST_DATA.csv', mode='w', newline='') as csv_file:
        fieldnames = ['DATE', 'AREA_CODE', 'AREA_NAME', 'TEMPERATURE', 'HUMIDITY', 'PRESSURE', 'WIND', 'RAIN', 'CLOUD', 'SNOW', 'TH_INDEX']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()

        for i in range (0,DATE.__len__()):
                writer.writerow(
                        {'DATE': DATE[i],
                         'AREA_CODE': AREA_CODE[i],
                         'AREA_NAME': AREA_NAME[i],
                         'TEMPERATURE': TEMPERATURE[i],
                         'HUMIDITY': HUMIDITY[i],
                         'PRESSURE': PRESSURE[i],
                         'WIND': WIND[i],
                         'RAIN': RAIN[i],
                         'CLOUD': CLOUD[i],
                         'SNOW': SNOW[i],
                         'TH_INDEX': PAIN[i]
                         })
                # print(HUMIDITY[i])
                
    print('Writing CSV with DATE -- Completed')
```

```python
def writeCsvForLearning():
    with open('SECOND_DATA.csv', mode='w', newline='') as csv_file:
        fieldnames = ['TEMPERATURE', 
                      'HUMIDITY', 
                      'PRESSURE', 
                      'WIND', 
                      'RAIN', 
                      'CLOUD', 
                      'SNOW', 
                      'TH_INDEX']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()

        for i in range (0,DATE.__len__()):
                writer.writerow(
                        {'TEMPERATURE': TEMPERATURE[i],
                         'HUMIDITY': HUMIDITY[i],
                         'PRESSURE': PRESSURE[i],
                         'WIND': WIND[i],
                         'RAIN': RAIN[i],
                         'CLOUD': CLOUD[i],
                         'SNOW': SNOW[i],
                         'TH_INDEX': PAIN[i]
                         })
                
    print('Writing CSV without DATE -- Completed')
```

```python
def writeCsvForLearningWithDate():
    with open('FINAL_DATA.csv', mode='w', newline='') as csv_file:
        fieldnames = ['DATE',
                      'MONTH',
                      'DAY',
                      'HOUR',
                      'TEMPERATURE', 
                      'HUMIDITY', 
                      'PRESSURE', 
                      'WIND', 
                      'RAIN', 
                      'CLOUD', 
                      'SNOW', 
                      'TH_INDEX']
        
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        
        
        for i in range (0,DATE.__len__()):
            if checkNotError(DATE[i],TEMPERATURE[i],HUMIDITY[i],PRESSURE[i]) == True:
                writer.writerow(
                        {'DATE':NEWNEWDATE[i],
                        'MONTH': MONTH[i],
                         'DAY': DAY[i],
                         'HOUR': HOUR[i],
                         'TEMPERATURE': TEMPERATURE[i],
                         'HUMIDITY': HUMIDITY[i],
                         'PRESSURE': PRESSURE[i],
                         'WIND': WIND[i],
                         'RAIN': RAIN[i],
                         'CLOUD': CLOUD[i],
                         'SNOW': SNOW[i],
                         'TH_INDEX': PAIN[i]
                         })
        
    print('Writing CSV with DATE -- Completed')
```

```python
def testCSV():
    with open('TEST_DATA.csv', mode='w', newline='') as csv_file:
        fieldnames = [
                      'MONTH',
                      'DAY',
                      'HOUR',
                      'TEMPERATURE', 
                      'HUMIDITY', 
                      'PRESSURE', 
                      'WIND', 
                      'RAIN', 
                      'CLOUD', 
                      'SNOW', 
                      'TH_INDEX']
        
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        
        
        for i in range (0,DATE.__len__()):
            if checkNotError(DATE[i],TEMPERATURE[i],HUMIDITY[i],PRESSURE[i]) == True:
                writer.writerow(
                        {
                        'MONTH': MONTH[i],
                         'DAY': DAY[i],
                         'HOUR': HOUR[i],
                         'TEMPERATURE': TEMPERATURE[i],
                         'HUMIDITY': HUMIDITY[i],
                         'PRESSURE': PRESSURE[i],
                         'WIND': WIND[i],
                         'RAIN': RAIN[i],
                         'CLOUD': CLOUD[i],
                         'SNOW': SNOW[i],
                         'TH_INDEX': PAIN[i]
                         })
        
    print('Writing TEST CSV with DATE -- Completed')
```

```python
def checkNotError(a,b,c,d):
    if a!=None and b!=0 and c!=0 and d!=0:
        return True
    else:
        return False
```

```python
def makeCSVforWeka():
    with open('LEARNINGWEKA_DATA.csv', mode='w', newline='') as csv_file:
        fieldnames = ['DATE',
                      'TEMPERATURE', 
                      'HUMIDITY', 
                      'PRESSURE', 
                      'WIND', 
                      'RAIN', 
                      'CLOUD', 
                      'SNOW', 
                      'TH_INDEX']
        
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        
        for i in range (0,DATE.__len__()):
                if checkNotError(DATE[i],TEMPERATURE[i],HUMIDITY[i],PRESSURE[i]) == True:
                    writer.writerow(
                        {'DATE':DATE[i],
                         'TEMPERATURE': TEMPERATURE[i],
                         'HUMIDITY': HUMIDITY[i],
                         'PRESSURE': PRESSURE[i],
                         'WIND': WIND[i],
                         'RAIN': RAIN[i],
                         'CLOUD': CLOUD[i],
                         'SNOW': SNOW[i],
                         'TH_INDEX': PAIN[i]
                     })
        
    print('Writing CSV for WEKA -- Completed')
```

```python
def makeCSVforWekaWithoutYear():
    with open('LEARNINGWEKA_DATA_2.csv', mode='w', newline='') as csv_file:
        fieldnames = ['DATE',
                      'TEMPERATURE', 
                      'HUMIDITY', 
                      'PRESSURE', 
                      'WIND', 
                      'RAIN', 
                      'CLOUD', 
                      'SNOW', 
                      'TH_INDEX']
        
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        
        for i in range (0,DATE.__len__()):
                if DATE[i]!=None and TEMPERATURE[i]!=0 and HUMIDITY[i]!=0 and PRESSURE[i]!=0:
                    writer.writerow(
                        {'DATE':NEWDATE[i],
                         'TEMPERATURE': TEMPERATURE[i],
                         'HUMIDITY': HUMIDITY[i],
                         'PRESSURE': PRESSURE[i],
                         'WIND': WIND[i],
                         'RAIN': RAIN[i],
                         'CLOUD': CLOUD[i],
                         'SNOW': SNOW[i],
                         'TH_INDEX': PAIN[i]
                     })
        
    print('Writing CSV without Year for WEKA -- Completed')
```

```python
def main():
    
    print('Start')
    print()
    
    readCsv() 
    writeCsv()
    writeCsvForLearning()
    writeCsvForLearningWithDate()
    makeCSVforWeka()
    makeCSVforWekaWithoutYear()
    testCSV()
    
    print()
    print("All -- Done")
                  
if __name__ == '__main__':
    main()
```
